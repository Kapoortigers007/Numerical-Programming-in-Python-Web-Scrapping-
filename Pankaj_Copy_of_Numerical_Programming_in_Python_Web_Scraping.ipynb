## **Scrapping Movies Data**

````python
# Specifying the URL from which movies related data will be fetched
headers = {

    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'

}
url='https://www.justwatch.com/in/movies?release_year_from=2000'
# Sending an HTTP GET request to the URL
response=requests.get(url, headers= headers)
# Parsing the HTML content using BeautifulSoup with the 'html.parser'
soup=BeautifulSoup(response.text,'html.parser')
# Printing the prettified HTML content
print(soup.prettify())
````

## **Fetching Movie URL's**

````python
# Write Your Code here
url_list=[]
for x in soup.find_all('a', class_="title-list-grid__item--link"):
 url_list.append('https://www.justwatch.com'+x['href'])
print(len(url_list))
print(url_list)
````

## **Scrapping Movie Title**

````python
# Write Your Code here
import time
movie_title = []
for url in url_list:
 try:
  response = requests.get(url, headers = headers)
  soup = BeautifulSoup(response.text,'html.parser')
  title=soup.find_all('h1')[0].text
 except:
  title='NA'
 movie_title.append(title)
 #time.sleep(1)
print(movie_title)
print(len(movie_title))



````

## **Scrapping release Year**

````python
# Write Your Code here
import time
movie_year=[]
for url in url_list:
 try:
   response=requests.get(url,headers=headers)
   soup=BeautifulSoup(response.text,'html.parser')
   year=soup.find_all('span',class_='release-year')[0].text.strip('()')
 except:
   year='NA'
 movie_year.append(year)
 #time.sleep(1)
print(movie_year)
print(len(movie_year))

````

## **Scrapping Genres**

````python
# Write Your Code here

import requests
from bs4 import BeautifulSoup

movie_genre = []

for url in url_list:
    genre = 'NA'  # default fallback
    try:
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text, 'html.parser')

        blocks = soup.find_all('div', class_='poster-detail-infos')

        for block in blocks:
            heading = block.find('h3', class_='poster-detail-infos__subheading')

            if heading and heading.text.strip() == 'Genres':
                value_div = block.find('div', class_='poster-detail-infos__value')

                if value_div:
                    span = value_div.find('span')

                    if span:
                        genre = span.text.strip()
                        break  # Stop after first matching block

    except Exception as e:
        print(f"Error on {url}: {e}")
        genre = 'NA'

    movie_genre.append(genre)

print(len(movie_genre))
print(movie_genre)
````

## **Scrapping IMBD Rating**

````python
# Write Your Code here
import time
import requests
from bs4 import BeautifulSoup

movie_imdbrating = []

for url in url_list:
    imdbrating = 'NA'  # default value
    try:
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text, 'html.parser')

        imdbratings = soup.find_all('span', class_='imdb-score')

        if imdbratings:
            imdbrating = imdbratings[0].text.strip().split()[0]

    except Exception as e:
        print(f"Error on {url}: {e}")
        imdbrating = 'NA'

    movie_imdbrating.append(imdbrating)
    # time.sleep(1)

print(len(movie_imdbrating))
print(movie_imdbrating)
````

## **Scrapping Runtime/Duration**

````python
# Write Your Code here
import requests
from bs4 import BeautifulSoup

movie_runtime = []

for url in url_list:
    runtime = 'NA'  # Default value
    try:
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find all divs with the correct class
        runtime_divs = soup.find_all('div', class_='title-detail-hero-details__item')

        for div in runtime_divs:
            text = div.get_text(strip=True)
            if 'min' in text or 'h' in text:  # crude check for runtime pattern
                runtime = text
                break  # stop after finding the first matching runtime

    except Exception as e:
        print(f"Error while processing {url}: {e}")
        runtime = 'NA'

    movie_runtime.append(runtime)

print(len(movie_runtime))
print(movie_runtime)
````

## **Scrapping Age Rating**

````python
# Write Your Code here
import requests
from bs4 import BeautifulSoup

movie_agerating = []

for url in url_list:
    age_rating = 'NA'  # Default fallback
    try:
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find all <h3> section headings
        for h3 in soup.find_all('h3', class_='poster-detail-infos__subheading'):
            if h3.text.strip() == 'Age rating':
                # Find the corresponding value container
                parent_div = h3.find_next_sibling('div', class_='poster-detail-infos__value')

                if parent_div:
                    age_rating = parent_div.text.strip()
                    break  # Stop once found

    except Exception as e:
        print(f"Error processing {url}: {e}")
        age_rating = 'NA'

    movie_agerating.append(age_rating)

print(len(movie_agerating))
print(movie_agerating)
````

## **Fetching Production Countries Details**

````python
# Write Your Code here
import requests
from bs4 import BeautifulSoup

movie_productioncountries = []

for url in url_list:
    production_country = 'NA'
    try:
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text, 'html.parser')

        for h3 in soup.find_all('h3', class_='poster-detail-infos__subheading'):
            if h3.text.strip() == 'Production country':
                next_div = h3.find_next_sibling('div', class_='poster-detail-infos__value')

                if next_div:
                    production_country = next_div.text.strip()
                    break

    except Exception as e:
        print(f"Error processing {url}: {e}")
        production_country = 'NA'

    movie_productioncountries.append(production_country)

print(len(movie_productioncountries))
print(movie_productioncountries)
````

## **Fetching Streaming Service Details**

````python
# Write Your Code here
import requests
from bs4 import BeautifulSoup

all_streaming_services = []

for url in url_list:
    services = []  # streaming services per URL
    try:
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text, 'html.parser')

        provider_icons = soup.find_all('img', {'data-testid': 'provider-icon-override'})

        for icon in provider_icons:
            service_name = icon.get('alt') or icon.get('title')
            if service_name:
                services.append(service_name)

    except Exception as e:
        print(f"Error processing {url}: {e}")
        services = ['NA']

    if not services:
        services = ['NA']

    all_streaming_services.append(services)

flat_streaming_services = [
    item for sublist in all_streaming_services for item in sublist
]

print(len(flat_streaming_services))
print(flat_streaming_services)
````

## **Task 2 :- Data Filtering & Analysis**

````python
# For Movies and TV shows release and start date between the Release year
start_year = 2005 # Converted to integer
end_year = 2024   # Converted to integer
min_rating = 7.0
# Ensure 'Release year' column is numeric. Replace 'NaN' string with actual NaN and then convert.
# 'errors="coerce"' will turn any non-numeric values (like the string 'NaN' that might be scraped) into actual NaN.
import pandas as pd
import numpy as np

# Filters
start_year = 2005   # Converted to integer
end_year = 2024     # Converted to integer
min_rating = 7.0

# Ensure 'release_year' column is numeric
# Replace string 'NA' / 'NaN' with actual NaN
final_df['release_year'] = final_df['release_year'].replace(['NA', 'NaN'], np.nan)

# Convert release_year to numeric safely
final_df['release_year'] = pd.to_numeric(
    final_df['release_year'],
    errors='coerce'   # Non-numeric values become NaN
)

# Ensure 'imdb_rating' is numeric (recommended)
final_df['imdb_rating'] = final_df['imdb_rating'].replace(['NA', 'NaN'], np.nan)
final_df['imdb_rating'] = pd.to_numeric(
    final_df['imdb_rating'],
    errors='coerce'
)

# Filter Movies & TV Shows by year range and minimum IMDb rating
filtered_df = final_df[
    (final_df['release_year'] >= start_year) &
    (final_df['release_year'] <= end_year) &
    (final_df['imdb_rating'] >= min_rating)
]

# Output
filtered_df.head()
````

````python
# Prepare Genre Data from final_df
from collections import Counter
import matplotlib.pyplot as plt

# Drop missing genres
genre_series = final_df['genre'].dropna()

all_genres = []

for genres in genre_series:
    if genres == 'NA':
        continue
    for g in genres.split(','):
        all_genres.append(g.strip())

# Get Top 5 Genres
genre_counts = Counter(all_genres)

top_5 = genre_counts.most_common(5)

top_5_genres_names = [g[0] for g in top_5]
top_5_genres_counts = [g[1] for g in top_5]

print(top_5_genres_names)
print(top_5_genres_counts)

# Lets visualize
plt.figure(figsize=(10, 6))
plt.bar(top_5_genres_names, top_5_genres_counts)

plt.xlabel('Genre')
plt.ylabel('Number of Movies and TV Shows')
plt.title('Top 5 Genres with the Most Movies and TV Shows')

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
````

````python
# Extracting genres and joining them into a single string

from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd

# Drop NaN first
genres_series = final_df['genre'].dropna()

# Convert to string
genres_series = genres_series.astype(str)

# Remove 'NA' and empty strings
genres_series = genres_series[genres_series != 'NA']

# Split multiple genres and flatten
all_genres = []
for genres in genres_series:
    for g in genres.split(','):
        g = g.strip()
        if g:
            all_genres.append(g)

# Join all genres into one string
genre_text = " ".join(all_genres)

# Generate WordCloud safely
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white'
).generate(genre_text)

# Display WordCloud
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Genre WordCloud (Movies & TV Shows)')
plt.show()
````

## **Finding Predominant Streaming Service**

````python
# Finding Predominant Streaming Service
from collections import Counter
import pandas as pd

# Drop missing values
services_series = final_df['streaming_service'].dropna()

all_services = []

for services in services_series:
    # Skip missing or invalid values
    if services in ['NA', '', None]:
        continue

    # Split multiple services (comma-separated)
    for s in str(services).split(','):
        s = s.strip()
        if s:
            all_services.append(s)

print("Total service entries:", len(all_services))

# Counting Streaming Services
service_counts = Counter(all_services)

# Convert to DataFrame
streaming_service_df = (
    pd.DataFrame(service_counts.items(),
                 columns=['Streaming Service', 'Count'])
    .sort_values(by='Count', ascending=False)
    .reset_index(drop=True)
)

# Display top services
print(streaming_service_df.head())

# Predominant Streaming Service
if not streaming_service_df.empty:
    predominant_service = streaming_service_df.iloc[0]
    print(
        f"Predominant Streaming Service: "
        f"{predominant_service['Streaming Service']} "
        f"with {predominant_service['Count']} titles"
    )
else:
    print("No streaming service data available.")
````
